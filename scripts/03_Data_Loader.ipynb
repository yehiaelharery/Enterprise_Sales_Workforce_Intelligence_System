```python
from sqlalchemy import create_engine
import pandas as pd

# Create engine using your existing connection details
engine = create_engine("postgresql://postgres:valeo2026@localhost:5432/enterprise_db")

# Load CSVs
regions_df = pd.read_csv(r"C:\Users\Lenovo\Documents\Enterprise_Sales_Workforce_Intelligence_System\data\regions.csv")
customers_df = pd.read_csv(r"C:\Users\Lenovo\Documents\Enterprise_Sales_Workforce_Intelligence_System\data\customers.csv")
products_df = pd.read_csv(r"C:\Users\Lenovo\Documents\Enterprise_Sales_Workforce_Intelligence_System\data\products.csv")
employees_df = pd.read_csv(r"C:\Users\Lenovo\Documents\Enterprise_Sales_Workforce_Intelligence_System\data\employees.csv")
sales_df = pd.read_csv(r"C:\Users\Lenovo\Documents\Enterprise_Sales_Workforce_Intelligence_System\data\sales.csv")


# Insert into PostgreSQL (IMPORTANT: append, not replace)
regions_df.to_sql("regions", engine, if_exists="append", index=False)
products_df.to_sql("products", engine, if_exists="append", index=False)
customers_df.to_sql("customers", engine, if_exists="append", index=False)
employees_df.to_sql("employees", engine, if_exists="append", index=False)
sales_df.to_sql("sales", engine, if_exists="append", index=False)


print("✅ Data inserted successfully into enterprise_db!")
```
